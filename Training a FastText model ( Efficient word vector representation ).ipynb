## Efficient leanrning of word representation

```python
from gensim.models.fasttext import FastText
```

```python
word_punctuation_tokenizer = nltk.WordPunctTokenizer()
word_tokenized_corpus = [word_punctuation_tokenizer.tokenize(sent) for sent in df['Text']]
word_tokenized_corpus[:5]
```

```python
embedding_size = 100
window_size = 5
min_word = 5
down_sampling = 1e-2
```

```python
%%time
ft_model = FastText(word_tokenized_corpus,
                      size=embedding_size,
                      window=window_size,
                      min_count=min_word,
                      sample=down_sampling,
                      sg=1,
                      iter=100)
```

```python
print(ft_model.wv['prema'])
```

```python
ft_model.wv.most_similar(positive = ['not'] , topn = 3)
```

```python
ft_model.wv.most_similar(negative = ['not'] , topn = 3)[0:]
```

```python
dic_of_similar = {word:[sim_words[0] for sim_words in ft_model.wv.most_similar([word],topn=5)] for word in ['share','interest','want','evening']}
for i,j in dic_of_similar.items():
    print(i , ":" , j)
```

```python
for word in ['share','interest','want','evening']:
    print(word)
```

```python
all_similar_words = sum([[k] + v for k, v in dic_of_similar.items()], [])
all_similar_words
```
